```python
def skewness(data):
    n = len(data)
    mean = sum(data) / n
    
    numerator = sum((x - mean)**3 for x in data) / n
    variance = sum((x - mean)**2 for x in data) / n
    denominator = variance ** 1.5
    
    return numerator / denominator


def kurtosis(data):
    n = len(data)
    mean = sum(data) / n
    
    numerator = sum((x - mean)**4 for x in data) / n
    variance = sum((x - mean)**2 for x in data) / n
    denominator = variance ** 2
    
    return numerator / denominator
```

```python
right_skewed = [1,2,2,3,3,3,4,4,10,15]
symmetric = [1,2,3,4,5,4,3,2,1]

print(skewness(right_skewed))
print(skewness(symmetric))
print(kurtosis(right_skewed))
print(kurtosis(symmetric))
```
