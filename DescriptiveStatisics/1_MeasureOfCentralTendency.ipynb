```python
import numpy as np
from scipy import stats

a = [2,4,1,5,63,3,64,23,12,240,9,32,5,2,6,3,87,43,98,54,6,1,53,32]

a = np.array(a , dtype=float)

print(f"Mean: {np.mean(a).round(2)}")
print(f"Median: {np.median(a)}")
print(f"Mode: {stats.mode(a)}")
```
