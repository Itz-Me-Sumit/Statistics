```python
class MSD:

    def __init__(self):
        pass
    

    def Range(self , arr):
        
        return np.max(arr) - np.min(arr)


    def Percentile(self , arr , percentile):

        arr = np.array(arr , dtype=float)

        sorted_arr = sorted(arr)
        n = len(arr)
        percentile /=100

        idx = (n-1) * percentile

        base_idx = int(idx)
        increment = idx - base_idx

        diff = sorted_arr[base_idx + 1] - sorted_arr[base_idx]

        result = sorted_arr[base_idx] +  ( increment * diff)
        
        return result

    def IQR(self , arr):
        
        q3 = self.Percentile(arr , 75)
        q1 = self.Percentile(arr , 25)

        return q3 - q1

    def var(self , arr):

        NumSum = 0

        for num in arr:
            NumSum+=num

        mean = NumSum/len(arr)

        non_avg_var = 0

        for num in arr:
            sq_deviation = (num-mean)**2
            non_avg_var += sq_deviation

        variance = non_avg_var/len(arr)
        return variance

    def std(self , arr):
        var = self.var(arr)
        return var ** 0.5

    def outliers(self , arr):
        
        arr = np.array(arr , dtype=float)

        q1 = self.Percentile(arr , 25)
        q3 = self.Percentile(arr , 75)

        iqr = q3-q1


        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr

        outlier = arr[ (arr < lower) | (arr>upper) ]

        return outlier if len(outlier)>0 else "No Outliers Present"  


        
```

```python
msd = MSD()

q1 = msd.Percentile(a , 25)
print(f"Q1 : {q1}")

q3 = msd.Percentile(a , 75)
print(f"Q3 : {q3}")

iqr = msd.IQR(a)
print(f"IQR : {iqr}")

rng = msd.Range(a)
print(f"Range : {rng}")

variance = msd.var(a)
print(f"Variance : {variance:.2f}")

std = msd.std(a)
print(f"Standard Deviation : {std:.3f}")

outlier_a = msd.outliers(a)
print(f"Outliers : {outlier_a}")

outlier_b = msd.outliers(b)
print(f"Outliers : {outlier_b}")
```
