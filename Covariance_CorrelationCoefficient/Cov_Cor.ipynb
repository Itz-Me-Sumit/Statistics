```python
import numpy as np
```

```python
class Relationship:

    def __init__(self):
        pass

    def variance(self , X):

        x = np.array(X , dtype=float)

        return (
            (((x - x.mean())**2).sum())/len(x)
        )

    def std(self , X):

        x = np.array(X , dtype=float)

        var_x = self.variance(x)

        return var_x ** 0.5


    def covariance(self , X , Y):

        x = np.array(X , dtype=float)
        y = np.array(Y , dtype=float)

        if not len(x) == len(y):
            raise ValueError("Length of both Array much be same")
        else:
            mean_x = x.mean()
            mean_y = y.mean()

            deviation_x = x - mean_x
            deviation_y = y - mean_y

            x_y_multi = deviation_x * deviation_y 
            
            return x_y_multi.sum()/len(x)
        

    def correlation(self , X , Y):

        x = np.array(X , dtype=float)
        y = np.array(Y , dtype=float)

        cov_xy = self.covariance(x , y)
        std_x = self.std(x)
        std_y = self.std(y)

        return (
            cov_xy/(
                std_x * std_y
            )
        )
```

```python
r = Relationship()

X = [2, 4, 5, 7, 8, 10, 12, 14, 15, 18]   # study hours
Y = [50, 55, 58, 65, 70, 75, 80, 88, 90, 95]   # marks

covariance = r.covariance(X , Y)
correlation = r.correlation(X , Y)

print(f"Covariance: {covariance:.3f}")
print(f"correlation: {correlation:.3f}")
```
