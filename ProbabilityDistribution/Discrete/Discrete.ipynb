```python
import random
import math
import numpy as np
import pandas as pd
```

```python
from abc import ABC , abstractmethod
```

```python
class BaseDistribution(ABC):


    @abstractmethod
    def pmf(self, x):
        raise NotImplementedError

    @property
    @abstractmethod
    def support(self):
        raise NotImplementedError

    def expected_value(self):
        e = 0
        for x in self.support:
            e += x * self.pmf(x)
        return e

    def variance(self):
        mean = self.expected_value()
        ex2 = 0
        for x in self.support:
            ex2 += (x**2) * self.pmf(x)
        return ex2 - mean**2

    def std(self):
        return (self.variance()) ** 0.5
```

```python
class UniformDiscrete(BaseDistribution):

    def __init__(self , start , end):
        self.start = start
        self.end = end
        self.n = self.end - self.start + 1

        if self.start > self.end:
            raise ValueError("start must be <= end")

    @property
    def support(self):
        return range(self.start , self.end+1)

    def pmf(self , x):
        if x < self.start or x > self.end:
            return 0
        return 1/self.n


class Bernoulli(BaseDistribution):

    def __init__(self , p):
        if not(0 <= p <= 1):
            raise ValueError("Probability must lie in [0,1]")
        self.p = p


    def pmf(self , x):
        if x == 1:
            return self.p
        elif x == 0:
            return (1-self.p)
        else:
            return 0

    @property
    def support(self):
        return [0,1]

```

```python
class Binomial(BaseDistribution):

    def __init__(self , n , p):
        

        if not (0<=p<=1):
            raise ValueError("Probability Range must be [0,1]")
        if n < 0:
            raise ValueError("n can't be negative")

        self.p = p
        self.n = n

    @property
    def support(self):
        return range(0 , self.n+1)
    
    def pmf(self , x):

        if x not in self.support:
            return 0
        
        return (
            math.comb(self.n , x)
            * (self.p ** x)
            * ( (1-self.p)**(self.n - x) )
        )

class Poisson(BaseDistribution):

    def __init__(self , lmda):
        if lmda <=0:
            raise ValueError("Lambda Must Be Positive")
        self.lmda = lmda

    @property
    def support(self):
        return range(0, int(self.lmda * 10))


    def pmf(self , x):

        return(
            math.exp(-self.lmda)
            * self.lmda ** x
            / math.factorial(x)
        )
```

```python
class Discrete:

    def __init__(self):
        pass

    @staticmethod
    def Uniform(start , end):

        return UniformDiscrete(start , end)


    @staticmethod
    def Bernoulli(p):
        
        return Bernoulli(p)

    @staticmethod
    def Binomial(n , p):

        return Binomial(n,p)

    @staticmethod
    def Poisson(lmda):

        return Poisson(lmda)
```
