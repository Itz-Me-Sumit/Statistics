## Pure Python

```python
def Factorial(n):

    if(n<0):
        return -1

    if n < 1:
        return 1
    return n * Factorial(n-1)
```

```python
def Permutation(n,r):

    if(n<r):
        return 0

    return Factorial(n)//Factorial(n-r)
```

```python
def Combination(n,r):
    
    if(n<r):
        return 0

    return Factorial(n) // ( Factorial(r) * Factorial(n-r) ) 
```

```python
print(Permutation(5,3))
print(Combination(10,5))
```
